```python
import numpy as np
import matplotlib.pyplot as plt

# データの準備
x=np.arange(0.0,2.5,0.05)*np.pi
s=np.sin(x)
c=np.cos(x)

# プロット
plt.subplot(2, 1, 1)  # 2行1列の1番目のプロット領域を指定
plt.plot(x,s,label="sin")
plt.plot(x,c,label="cos")
plt.title('sin/cos functions')
plt.legend() #凡例(上記のlabel=の値)表示


plt.subplot(2, 1, 2)  # 2行1列の2番目のプロット領域を指定
plt.plot(x,s,"o",label="sin")
plt.plot(x,c,"o",label="cos")
plt.title('sin/cos functions(dot)')
plt.legend() #凡例(上記のlabel=の値)表示

# グラフの表示
plt.tight_layout()  # プロット間のスペースを自動調整
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
    return np.maximum(0, x)

def LeakyReLU(x, alpha):
    return np.maximum(alpha * x, x)

# 入力値の範囲を定義
x = np.linspace(-5, 5, 100)

# ReLU関数を適用した出力値を計算
y_relu = ReLU(x)

# Leaky ReLU関数を適用した出力値を計算
alpha = 0.2  # リークする値
y_leaky_relu = LeakyReLU(x, alpha)

# グラフを描画
plt.plot(x, y_relu, label='ReLU')
plt.plot(x, y_leaky_relu, label='Leaky ReLU')
plt.xlabel('x')
plt.ylabel('Activation')
plt.title('ReLU vs Leaky ReLU')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['red', 'blue', 'yellow']

y_total = [y1_val + y2_val for y1_val, y2_val in zip(y1, y2)]

ax.bar(x, y_total, tick_label=labels, label='y1')  # 【イ】
ax.bar(x, y2, label='y2')  # 【ウ】
ax.legend()

plt.show()
```
