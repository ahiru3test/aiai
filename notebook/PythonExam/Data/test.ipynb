```python
import numpy as np
import matplotlib.pyplot as plt

# データの準備
x=np.arange(0.0,2.5,0.05)*np.pi
s=np.sin(x)
c=np.cos(x)

# プロット
plt.subplot(2, 1, 1)  # 2行1列の1番目のプロット領域を指定
plt.plot(x,s,label="sin")
plt.plot(x,c,label="cos")
plt.title('sin/cos functions')
plt.legend() #凡例(上記のlabel=の値)表示


plt.subplot(2, 1, 2)  # 2行1列の2番目のプロット領域を指定
plt.plot(x,s,"o",label="sin")
plt.plot(x,c,"o",label="cos")
plt.title('sin/cos functions(dot)')
plt.legend() #凡例(上記のlabel=の値)表示

# グラフの表示
plt.tight_layout()  # プロット間のスペースを自動調整
plt.show()
```

### MatPlotLib
