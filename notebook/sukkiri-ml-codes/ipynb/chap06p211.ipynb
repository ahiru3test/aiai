```python
import matplotlib.pyplot as plt
import numpy as np
 
### y = ax + b のデータを作成
a = 3
b = 2
x = np.random.rand(100, 1)  # 0 〜 1 までの乱数を 100 個つくる
x = x*4 - 2                 # 値の範囲を -2 〜 2 に変更
 
y = a*x + b
 
### 学習
from sklearn import linear_model
 
model = linear_model.LinearRegression()
model.fit(x, y)
 
### 係数、切片を表示
print('係数', model.coef_)
print('切片', model.intercept_)
 
### グラフ表示
plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
### y = ax + b のデータを作成
a = 3
b = 2
x = np.random.rand(100, 1)  # 0 〜 1 までの乱数を 100 個つくる
x = x*4 - 2                 # 値の範囲を -2 〜 2 に変更
 
y = a*x + b
 
# 標準正規分布(平均0, 偏差1)の乱数を加える
y += np.random.randn(100,1)
 
### 学習
from sklearn import linear_model
 
model = linear_model.LinearRegression()
model.fit(x, y)
 
### 係数、切片を表示
print('係数', model.coef_)
print('切片', model.intercept_)
 
### グラフ表示
plt.scatter(x, y)
plt.scatter(x, model.predict(x), marker="+")
plt.show()
```
