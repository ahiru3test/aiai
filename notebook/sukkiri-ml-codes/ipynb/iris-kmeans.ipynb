```python
# -*- coding: utf-8 -*-
import os
os.environ["OMP_NUM_THREADS"] = '1'
# https://stackoverflow.com/questions/69596239/how-to-avoid-memory-leak-when-dealing-with-kmeans-for-example-in-this-code-i-am
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn import datasets
 
# iris データをロード
iris = datasets.load_iris()
data = iris["data"]
 
# 初期中心点を定義
init_centers=np.array([
       [4,2.5,3,0],
       [5,3  ,3,1],
       [6,4  ,3,2]])
 
# データ定義と値の取り出し
x_index = 1
y_index = 2
 
data_x=data[:,x_index]
data_y=data[:,y_index]
 
# グラフのスケールとラベルの定義
x_label = iris["feature_names"][x_index]
y_label = iris["feature_names"][y_index]
 
def show_result(cluster_centers,labels, ax):
    
    # cluster 0と中心点を描画
    ax.scatter(data_x[labels==0], data_y[labels==0],c='red' ,alpha=0.3,s=100, marker="o",label="cluster 0")
    ax.scatter(cluster_centers[0][x_index], cluster_centers[0][y_index],facecolors='white', edgecolors='red', s=300, marker="o")
 
     # cluster １と中心点を描画
    ax.scatter(data_x[labels==1], data_y[labels==1],c='green' ,alpha=0.3,s=100, marker="^",label="cluster 1")
    ax.scatter(cluster_centers[1][x_index], cluster_centers[1][y_index],facecolors='white', edgecolors='green', s=300, marker="^")
 
     # cluster と中心点を描画
    ax.scatter(data_x[labels==2], data_y[labels==2],c='blue' ,alpha=0.3,s=100, marker="*",label="cluster 2")
    ax.scatter(cluster_centers[2][x_index], cluster_centers[2][y_index],facecolors='white', edgecolors='blue', s=500, marker="*")
 
    # グラフのスケールと軸ラベルを設定
    ax.set_xlabel(x_label,fontsize='large')
    ax.set_ylabel(y_label,fontsize='large')
 
# matplotlib figure定義
fig = plt.figure(figsize=(8,8))
 
# 初期状態を表示
labels=np.zeros(len(data),dtype=np.int32)
ax = fig.add_subplot(3,2,1)
show_result(init_centers,labels, ax)
 
# https://cookie-box.hatenablog.com/entry/2017/06/11/224032
# max_iter : クラスタリングの最大反復数
# init : 各クラスターの初期中心位置
# n_init : クラスタリング回数
for i in range(1,6):
    model = cluster.KMeans(n_clusters=3,max_iter=1,init=init_centers, n_init=1).fit(data)
    labels = model.labels_
    init_centers=model.cluster_centers_
    ax = fig.add_subplot(3,2,i+1)
    show_result(init_centers,labels, ax)
 
plt.tight_layout()
plt.show()
```
